Выполнить расчёты при каком-то одном выбранном числе Куранта по монотонной схеме второго-третьего порядка аппроксимации, каждый из коэффициентов $\alpha^\nu_\mu$ которой нужно задавать в виде линейной комбинации 3 опорных схем -- одной схемы 3-го порядка аппроксимации и двух схем 2-го порядка аппроксимации, с ортогональными областями монотонности.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lmb = 0.05
L = 1 
T = 100 
dx = 1e-3 
dt = 1e-2 
s = lmb * dt / dx
N = int(T / dt)
M = int(L / dx)
u= np.zeros((N, M))
```

```python
for i in range(int(M/10)+1):
    u[0, int(M/10)+i] = i*10/M
    u[1, int(M/10)+i] = i*10/M

for i in range(int(M/10)):
    u[0,int(M/10)+ int(M/5) -i] = i*10/M
    u[1, int(M/10)+ int(M/5) -i] = i*10/M

plt.plot(u[0])
plt.plot(u[1])
plt.grid()
```

```python
u[0, int(M/10):int(M/10 * 2)] = u[1, int(M/10):int(M/10 * 2)] = 1
plt.plot(u[0])
plt.grid()
```

```python
for i in range(2, N):
    for j in range(2, M-1):
        u[i, j] = 0.8 * u[i-1, j-1] - 1/45* u[i-2, j+1] -2/9 * u[i-2, j-2] + 4/9* u[i-1, j]
```

```python
plt.plot(u[0], label='t = 0')
plt.plot(u[100], label='t = 1.0')
plt.plot(u[1000], label='t = 10.0')
plt.plot(u[5000], label='t = 50.0')
plt.legend()
plt.grid()
```
